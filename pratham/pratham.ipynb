```python
import pandas as pd
import matplotlib as plt
```

```python
bridges = pd.read_csv("../assets/bridges.csv", sep=",", header=0)
speed_humps = pd.read_csv("../assets/speed_humps.csv", sep=",", header=0)
print(speed_humps)
```

```python
res = speed_humps.groupby('neighborhood')['_id'].sum().sort_values(ascending=True)
# Create a bar plot
ax = res.plot(kind='line', title="Speed Bumps by Neighborhood", xlabel="", ylabel="Bumps")
# Get the x-axis tick labels
x_labels = res.index

ax.set_xticks(range(len(x_labels)))
ax.set_xticklabels(x_labels, rotation=90)
```
